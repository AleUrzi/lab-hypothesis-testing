**Challenge 1**

In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
#code here
dragon_pokemon = df[df["Type 1"]=="Grass"]["HP"]
grass_pokemon = df[df["Type 1"]=="Dragon"]["HP"]

#Set the hypothesis

#H0: Pokemons of type Dragon have, on average, more HP stats than Grass. 
#H1: mean HP grass pokemon > mean HP Dragon pokemon

#significance level = 0.05

st.ttest_ind(dragon_pokemon,grass_pokemon, equal_var=False, alternative = "greater" )

#p value > 0.05: H1 is rejected and H0 is confirmed: Dragon pokemon have indeed more average HP than grass pokemon
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
#code here

stats = df.groupby("Legendary")[["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]].agg("mean")
display(stats)

legend_pokemon = df[df["Legendary"] == True][["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]
not_legend_pokemon = df[df["Legendary"] == False][["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]

#Set the hypothesis

#H1: Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. 
#H0: opposite: no difference

#significance level = 0.05


# Perform independent t-tests for each stat
for stat in ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]:
    stat_test = st.ttest_ind(legend_pokemon[stat], not_legend_pokemon[stat], equal_var=False)
    print(f"Stat: {stat} : {stat_test}")

#p values < 0.05: H0 is rejected and H1 is confirmed: There is a difference betwe leg and non leg pokemon stats
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
def distance(df):
    # Assuming the DataFrame has 'longitude' and 'latitude' columns
    school_coords = (-118, 37)
    hospital_coords = (-122, 34)

    # Calculate distances
    df['distance_to_school'] = np.sqrt((df['longitude'] - school_coords[0]) ** 2 + (df['latitude'] - school_coords[1]) ** 2)
    df['distance_to_hospital'] = np.sqrt((df['longitude'] - hospital_coords[0]) ** 2 + (df['latitude'] - hospital_coords[1]) ** 2)
    
    return df[['distance_to_school', 'distance_to_hospital']]

display(distance(df))

close_to_school = df[df['distance_to_school'] < 0.5]["median_house_value"]
far_from_school = df[df['distance_to_school'] > 0.5]["median_house_value"]

close_to_hospital = df[df['distance_to_hospital'] < 0.5]["median_house_value"]
far_from_hospital = df[df['distance_to_hospital'] > 0.5]["median_house_value"]

# H0 : We posit that houses close to either a school or a hospital are more expensive.
#H1: Not more expensive


display(st.ttest_ind(close_to_school,far_from_school, equal_var=False, alternative = "greater"))
display(st.ttest_ind(close_to_hospital,far_from_hospital, equal_var=False, alternative = "less"))

print("Close to Hospital count:", close_to_hospital.count())
print("Far from Hospital count:", far_from_hospital.count())

#Houses close to school are NOT more expensive
#since there are no houses close to hospital, the t test failed
```
